```python
import requests
from bs4 import BeautifulSoup
from IPython.display import display, Markdown
from tabulate import tabulate

url = "https://www.zerozero.pt/jogo/2025-05-11-vitoria-sc-farense/10240205"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)
response.raise_for_status()
soup = BeautifulSoup(response.text, 'html.parser')

section = soup.select_one('#page_rightbar > div:nth-child(5)')

if not section:
    print("Sec√ß√£o de √°rbitros n√£o encontrada!")
    exit()

staff_lines = section.find_all('div', class_='staff_line')

arbitros_por_secao = {}

for staff in staff_lines:
    previous_section = staff.find_previous_sibling('div', class_='section')
    section_title = previous_section.get_text(strip=True) if previous_section else "Outro"

    photo_div = staff.find('div', class_='photo')
    photo_style = photo_div['style'] if photo_div else ""
    photo_url = ""
    if photo_style.startswith('background-image: url('):
        start = photo_style.find("url(") + 4
        end = photo_style.find(")", start)
        photo_path = photo_style[start:end].strip("'\"")
        photo_url = "https://www.zerozero.pt" + photo_path

    text_div = staff.find('div', class_='text')
    name_link = text_div.find('a') if text_div else None
    name = name_link.get_text(strip=True) if name_link else "Sem nome"
    profile_link = "https://www.zerozero.pt" + name_link['href'] if name_link else ""

    name_div = staff.find('div', class_='entity name')
    country_img = name_div.find('img') if name_div else None
    country = country_img['alt'] if country_img else ""

    spans = name_div.find_all('span') if name_div else []
    age = spans[0].get_text(strip=True) if len(spans) > 0 else ""
    association = spans[1].get_text(strip=True) if len(spans) > 1 else ""

    if section_title not in arbitros_por_secao:
        arbitros_por_secao[section_title] = []

    arbitros_por_secao[section_title].append([
        f'<img src="{photo_url}" width="80">' if photo_url else "",  # Foto
        name,                                                        # Nome
        association,                                                 # Associa√ß√£o
        country,                                                    # Pa√≠s
        age,                                                        # Idade
        profile_link                                                # Link
    ])

for secao, arbitros in arbitros_por_secao.items():
    display(Markdown(f"## {secao}\n"))
    tabela = tabulate(arbitros, headers=["Foto", "Nome", "Associa√ß√£o", "Pa√≠s", "Idade", "Link"], tablefmt="unsafehtml")
    display(Markdown(tabela))
```

```python
import requests
from bs4 import BeautifulSoup
import json
import time
import re

# Links dos jogos do Farense na temporada 2024/25
links_farense_2024_25 = [
    "https://www.zerozero.pt/jogo/2025-05-17-farense-santa-clara/10240229",
    "https://www.zerozero.pt/jogo/2025-05-11-vitoria-sc-farense/10240205",
    "https://www.zerozero.pt/jogo/2025-05-03-farense-fc-famalicao/10240191",
    "https://www.zerozero.pt/jogo/2025-04-26-gil-vicente-farense/10240167",
    "https://www.zerozero.pt/jogo/2025-04-18-farense-boavista/10240142",
    "https://www.zerozero.pt/jogo/2025-04-13-est-amadora-farense/10240147",
    "https://www.zerozero.pt/jogo/2025-04-07-farense-casa-pia-ac/10240115",
    "https://www.zerozero.pt/jogo/2025-04-02-benfica-farense/10240079",
    "https://www.zerozero.pt/jogo/2025-03-15-farense-sc-braga/10240093",
    "https://www.zerozero.pt/jogo/2025-03-08-estoril-praia-farense/10240047",
    "https://www.zerozero.pt/jogo/2025-03-02-farense-afs/10240067",
    "https://www.zerozero.pt/jogo/2025-02-23-fc-arouca-farense/10240009",
    "https://www.zerozero.pt/jogo/2025-02-16-farense-fc-porto/10240035",
    "https://www.zerozero.pt/jogo/2025-02-09-farense-nacional/10239995",
    "https://www.zerozero.pt/jogo/2025-02-02-sporting-farense/10239975",
    "https://www.zerozero.pt/jogo/2025-01-25-farense-rio-ave/10239961",
    "https://www.zerozero.pt/jogo/2025-01-19-moreirense-farense/10239933",
    "https://www.zerozero.pt/jogo/2025-01-14-farense-benfica/10593811",
    "https://www.zerozero.pt/jogo/2025-01-05-santa-clara-farense/10240228",
    "https://www.zerozero.pt/jogo/2024-12-29-farense-vitoria-sc/10240204",
    "https://www.zerozero.pt/jogo/2024-12-21-fc-famalicao-farense/10240190",
    "https://www.zerozero.pt/jogo/2024-12-14-farense-gil-vicente/10240166",
    "https://www.zerozero.pt/jogo/2024-12-08-boavista-farense/10240139",
    "https://www.zerozero.pt/jogo/2024-11-29-farense-est-amadora/10240145",
    "https://www.zerozero.pt/jogo/2024-11-23-fc-arouca-farense/10593786",
    "https://www.zerozero.pt/jogo/2024-11-09-casa-pia-ac-farense/10240114",
    "https://www.zerozero.pt/jogo/2024-11-02-farense-benfica/10240078",
    "https://www.zerozero.pt/jogo/2024-10-27-sc-braga-farense/10240092",
    "https://www.zerozero.pt/jogo/2024-10-20-ad-sanjoanense-farense/10517931",
    "https://www.zerozero.pt/jogo/2024-10-06-farense-estoril-praia/10240046",
    "https://www.zerozero.pt/jogo/2024-09-30-afs-farense/10240066",
    "https://www.zerozero.pt/jogo/2024-09-22-farense-fc-arouca/10240008",
    "https://www.zerozero.pt/jogo/2024-09-15-fc-porto-farense/10240034",
    "https://www.zerozero.pt/jogo/2024-09-01-nacional-farense/10239994",
    "https://www.zerozero.pt/jogo/2024-08-23-farense-sporting/10239974",
    "https://www.zerozero.pt/jogo/2024-08-17-rio-ave-farense/10239960",
    "https://www.zerozero.pt/jogo/2024-08-11-farense-moreirense/10239932",
]

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Lista para armazenar todos os dados
todos_jogos = []

for url in links_farense_2024_25:
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extrair informa√ß√µes do jogo
        info_div = soup.select_one("#page_header > div > div.info")
        if not info_div:
            print(f"‚ùå Div 'info' n√£o encontrada em {url}")
            continue
        
        # Data do jogo (extraindo da URL)
        url_date_match = re.search(r'/jogo/(\d{4}-\d{2}-\d{2})-', url)
        data_jogo = url_date_match.group(1) if url_date_match else "N/A"
        
        # Arbitro principal
        arbitro_link = info_div.select_one(".header a[href^='/arbitro']")
        arbitro = arbitro_link.text.strip() if arbitro_link else "N/A"
        
        # Campeonato
        campeonato_link = info_div.select_one(".header a[href^='/edicao']")
        campeonato = campeonato_link.text.strip() if campeonato_link else "N/A"
        
        # Jornada
        jornada_text = info_div.select_one(".header").text
        jornada_match = re.search(r'Jornada\s+(\d+)', jornada_text)
        jornada = jornada_match.group(1) if jornada_match else "N/A"
        
        # Equipas e resultado
        home_team = info_div.select_one(".home .team a").text.strip()
        away_team = info_div.select_one(".away .team a").text.strip()
        resultado = info_div.select_one(".score a").text.strip()
        
        # Criar dicion√°rio para armazenar detalhes deste jogo
        jogo_info = {
            "data": data_jogo,
            "url": url,
            "campeonato": campeonato,
            "jornada": jornada,
            "equipa_casa": home_team,
            "equipa_fora": away_team,
            "resultado": resultado,
            "arbitros": {}
        }
        
        # Extrair informa√ß√µes de todos os √°rbitros
        section = soup.select_one('#page_rightbar > div:nth-child(5)')
        
        if section:
            staff_lines = section.find_all('div', class_='staff_line')
            
            for staff in staff_lines:
                previous_section = staff.find_previous_sibling('div', class_='section')
                section_title = previous_section.get_text(strip=True) if previous_section else "Outro"
                
                text_div = staff.find('div', class_='text')
                name_link = text_div.find('a') if text_div else None
                nome = name_link.get_text(strip=True) if name_link else "Sem nome"
                
                name_div = staff.find('div', class_='entity name')
                spans = name_div.find_all('span') if name_div else []
                associacao = spans[1].get_text(strip=True) if len(spans) > 1 else ""
                
                # Armazenar informa√ß√µes dos √°rbitros no dicion√°rio do jogo
                if section_title not in jogo_info["arbitros"]:
                    jogo_info["arbitros"][section_title] = []
                    
                jogo_info["arbitros"][section_title].append({
                    "nome": nome,
                    "associacao": associacao
                })
        else:
            print(f"‚ùå Sec√ß√£o de √°rbitros n√£o encontrada em {url}")
        
        # Adicionar informa√ß√µes deste jogo √† lista principal
        todos_jogos.append(jogo_info)
        
        print(f"‚úÖ Processado: {data_jogo} - {home_team} {resultado} {away_team}")
        
    except Exception as e:
        print(f"‚ùó Erro ao processar {url}: {e}")
    
    # Pausa para n√£o sobrecarregar o servidor
    time.sleep(1.5)

# Criar o objeto JSON final
dados_json = {
    "temporada": "2024-25",
    "equipa": "Farense",
    "jogos": todos_jogos
}

# Salvar os dados no arquivo JSON
with open("arbitros_2024_25.json", "w", encoding="utf-8") as json_file:
    json.dump(dados_json, json_file, ensure_ascii=False, indent=4)

print(f"\n‚úÖ Dados de {len(todos_jogos)} jogos salvos com sucesso em 'arbitros_2024_25.json'")
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from collections import Counter, defaultdict
import plotly.io as pio
from plotly.subplots import make_subplots
import numpy as np

# Configura√ß√£o para exibir gr√°ficos no notebook/ambiente
pio.renderers.default = "notebook"

# Carregar os dados do JSON
try:
    with open("arbitros_2024_25.json", "r", encoding="utf-8") as file:
        dados = json.load(file)
except FileNotFoundError:
    print("Arquivo 'arbitros_2024_25.json' n√£o encontrado!")
    exit()

print(f"‚úÖ Dados carregados: {len(dados['jogos'])} jogos do {dados['equipa']} na temporada {dados['temporada']}")

# Fun√ß√µes auxiliares
def extrair_gols_do_resultado(resultado):
    """Extrai os gols marcados de um resultado no formato '2-1'"""
    if resultado == "N/A" or "-" not in resultado:
        return 0, 0
    partes = resultado.split("-")
    return int(partes[0]), int(partes[1])

def calcular_resultado(equipa, gols_casa, gols_fora, equipa_casa):
    """Determina se o resultado foi vit√≥ria, empate ou derrota para a equipa"""
    if gols_casa == gols_fora:
        return "Empate"
    elif (equipa_casa and equipa == equipa_casa and gols_casa > gols_fora) or \
         (not equipa_casa and equipa != equipa_casa and gols_fora > gols_casa):
        return "Vit√≥ria"
    else:
        return "Derrota"

# Organizar os dados em DataFrames
jogos_df = []

for jogo in dados['jogos']:
    gols_casa, gols_fora = extrair_gols_do_resultado(jogo['resultado'])
    
    # Verificar se o Farense jogou em casa ou fora
    farense_casa = jogo['equipa_casa'] == dados['equipa']
    
    # Resultado para o Farense
    resultado_farense = calcular_resultado(
        dados['equipa'], 
        gols_casa, 
        gols_fora, 
        jogo['equipa_casa']
    )
    
    # Gols marcados e sofridos pelo Farense
    gols_farense = gols_casa if farense_casa else gols_fora
    gols_sofridos = gols_fora if farense_casa else gols_casa
    
    # Adicionar √°rbitro principal ao DataFrame
    arbitro_principal = None
    arbitro_associacao = None
    
    if "√Årbitro" in jogo['arbitros'] and jogo['arbitros']["√Årbitro"]:
        arbitro_principal = jogo['arbitros']["√Årbitro"][0]['nome']
        arbitro_associacao = jogo['arbitros']["√Årbitro"][0]['associacao']
    
    # Adicionar VAR ao DataFrame (se existir)
    var_principal = None
    if "VAR" in jogo['arbitros'] and jogo['arbitros']["VAR"]:
        var_principal = jogo['arbitros']["VAR"][0]['nome']
    
    # Criar registro para o DataFrame
    jogo_dict = {
        'data': jogo['data'],
        'campeonato': jogo['campeonato'],
        'jornada': jogo['jornada'],
        'equipa_casa': jogo['equipa_casa'],
        'equipa_fora': jogo['equipa_fora'],
        'resultado': jogo['resultado'],
        'gols_casa': gols_casa,
        'gols_fora': gols_fora,
        'farense_casa': farense_casa,
        'resultado_farense': resultado_farense,
        'gols_farense': gols_farense,
        'gols_sofridos': gols_sofridos,
        'arbitro': arbitro_principal,
        'arbitro_associacao': arbitro_associacao,
        'var': var_principal
    }
    
    jogos_df.append(jogo_dict)

# Criar o DataFrame principal
df = pd.DataFrame(jogos_df)

# Converter a coluna 'data' para datetime
df['data'] = pd.to_datetime(df['data'])
df = df.sort_values('data')

# Para an√°lises temporais, extrair m√™s e ano
df['mes_ano'] = df['data'].dt.strftime('%Y-%m')

print(f"‚úÖ DataFrame criado com {len(df)} jogos")

# An√°lises e Estat√≠sticas
print("\n==== ESTAT√çSTICAS DOS √ÅRBITROS ====\n")

# 1. Contagem de jogos por √°rbitro
arbitros_jogos = df['arbitro'].value_counts().reset_index()
arbitros_jogos.columns = ['√Årbitro', 'Jogos']
print("Top 10 √°rbitros com mais jogos:")
print(arbitros_jogos.head(10))

# 2. Performance do Farense com cada √°rbitro
arbitro_performance = df.groupby('arbitro')['resultado_farense'].value_counts().unstack().fillna(0)
if 'Vit√≥ria' not in arbitro_performance.columns:
    arbitro_performance['Vit√≥ria'] = 0
if 'Empate' not in arbitro_performance.columns:
    arbitro_performance['Empate'] = 0
if 'Derrota' not in arbitro_performance.columns:
    arbitro_performance['Derrota'] = 0

arbitro_performance['Total'] = arbitro_performance['Vit√≥ria'] + arbitro_performance['Empate'] + arbitro_performance['Derrota']
arbitro_performance['Pontos'] = arbitro_performance['Vit√≥ria'] * 3 + arbitro_performance['Empate']
arbitro_performance['Aproveitamento'] = (arbitro_performance['Pontos'] / (arbitro_performance['Total'] * 3)) * 100

# Filtrar √°rbitros com pelo menos 2 jogos
arbitro_performance = arbitro_performance[arbitro_performance['Total'] >= 2].sort_values('Aproveitamento', ascending=False)

print("\nPerformance do Farense com diferentes √°rbitros (min. 2 jogos):")
print(arbitro_performance[['Vit√≥ria', 'Empate', 'Derrota', 'Total', 'Pontos', 'Aproveitamento']].round(1).head(10))

# 3. M√©dia de gols por jogo com cada √°rbitro
gols_por_arbitro = df.groupby('arbitro').agg({
    'gols_farense': ['mean', 'sum'],
    'gols_sofridos': ['mean', 'sum'],
    'arbitro': 'count'
}).reset_index()

gols_por_arbitro.columns = ['arbitro', 'media_gols_marcados', 'total_gols_marcados', 
                           'media_gols_sofridos', 'total_gols_sofridos', 'jogos']
gols_por_arbitro = gols_por_arbitro[gols_por_arbitro['jogos'] >= 2].sort_values('media_gols_marcados', ascending=False)

print("\nM√©dia de gols por jogo com diferentes √°rbitros (min. 2 jogos):")
print(gols_por_arbitro[['arbitro', 'media_gols_marcados', 'media_gols_sofridos', 'jogos']].round(2).head(10))

# 4. Estat√≠sticas por associa√ß√£o de √°rbitros
associacao_performance = df.groupby('arbitro_associacao')['resultado_farense'].value_counts().unstack().fillna(0)
if 'Vit√≥ria' not in associacao_performance.columns:
    associacao_performance['Vit√≥ria'] = 0
if 'Empate' not in associacao_performance.columns:
    associacao_performance['Empate'] = 0
if 'Derrota' not in associacao_performance.columns:
    associacao_performance['Derrota'] = 0

associacao_performance['Total'] = associacao_performance['Vit√≥ria'] + associacao_performance['Empate'] + associacao_performance['Derrota']
associacao_performance['Pontos'] = associacao_performance['Vit√≥ria'] * 3 + associacao_performance['Empate']
associacao_performance['Aproveitamento'] = (associacao_performance['Pontos'] / (associacao_performance['Total'] * 3)) * 100

# Filtrar associa√ß√µes com pelo menos 3 jogos
associacao_performance = associacao_performance[associacao_performance['Total'] >= 3].sort_values('Aproveitamento', ascending=False)

print("\nPerformance do Farense com diferentes associa√ß√µes de √°rbitros (min. 3 jogos):")
print(associacao_performance[['Vit√≥ria', 'Empate', 'Derrota', 'Total', 'Pontos', 'Aproveitamento']].round(1))

# -----------------------------------------
# GR√ÅFICOS 
# -----------------------------------------

# Criar diret√≥rio para gr√°ficos se n√£o existir
import os
if not os.path.exists("estatisticas_arbitros"):
    os.makedirs("estatisticas_arbitros")

# 1. Gr√°fico de barras - Top 10 √°rbitros com mais jogos
fig_top10 = px.bar(
    arbitros_jogos.head(10),
    x='√Årbitro',
    y='Jogos',
    title='Top 10 √Årbitros com Mais Jogos do Farense 2024/25',
    color='Jogos',
    color_continuous_scale='Blues',
    text='Jogos'
)
fig_top10.update_layout(xaxis_title='√Årbitro', yaxis_title='N√∫mero de Jogos')
fig_top10.write_html("estatisticas_arbitros/top10_arbitros_jogos.html")
fig_top10.write_image("estatisticas_arbitros/top10_arbitros_jogos.png", width=900, height=600)

# 2. Gr√°fico de barras empilhadas - Performance do Farense com cada √°rbitro (min. 2 jogos)
top_arbitros = arbitro_performance.head(10).reset_index()
fig_performance = px.bar(
    top_arbitros,
    x='arbitro',
    y=['Vit√≥ria', 'Empate', 'Derrota'],
    title='Performance do Farense com Diferentes √Årbitros (m√≠n. 2 jogos)',
    color_discrete_map={'Vit√≥ria': '#2ecc71', 'Empate': '#f1c40f', 'Derrota': '#e74c3c'},
    barmode='stack'
)
fig_performance.update_layout(xaxis_title='√Årbitro', yaxis_title='N√∫mero de Jogos')
fig_performance.write_html("estatisticas_arbitros/performance_por_arbitro.html")
fig_performance.write_image("estatisticas_arbitros/performance_por_arbitro.png", width=900, height=600)

# 3. Gr√°fico de aproveitamento de pontos por √°rbitro
fig_aproveitamento = px.bar(
    top_arbitros,
    x='arbitro',
    y='Aproveitamento',
    title='Aproveitamento de Pontos do Farense com Diferentes √Årbitros (%)',
    color='Aproveitamento',
    color_continuous_scale='RdYlGn',
    text=top_arbitros['Aproveitamento'].round(1).astype(str) + '%'
)
fig_aproveitamento.update_layout(xaxis_title='√Årbitro', yaxis_title='Aproveitamento (%)')
fig_aproveitamento.write_html("estatisticas_arbitros/aproveitamento_por_arbitro.html")
fig_aproveitamento.write_image("estatisticas_arbitros/aproveitamento_por_arbitro.png", width=900, height=600)

# 4. Gr√°fico de m√©dia de gols por jogo com cada √°rbitro
top_gols = gols_por_arbitro.head(10)
fig_gols = go.Figure()
fig_gols.add_trace(go.Bar(
    x=top_gols['arbitro'],
    y=top_gols['media_gols_marcados'],
    name='Gols Marcados',
    marker_color='#2ecc71'
))
fig_gols.add_trace(go.Bar(
    x=top_gols['arbitro'],
    y=top_gols['media_gols_sofridos'],
    name='Gols Sofridos',
    marker_color='#e74c3c'
))
fig_gols.update_layout(
    title='M√©dia de Gols por Jogo com Diferentes √Årbitros',
    xaxis_title='√Årbitro',
    yaxis_title='M√©dia de Gols',
    barmode='group'
)
fig_gols.write_html("estatisticas_arbitros/media_gols_por_arbitro.html")
fig_gols.write_image("estatisticas_arbitros/media_gols_por_arbitro.png", width=900, height=600)

# 5. Gr√°fico de Radar - Comportamento com Diferentes Associa√ß√µes
associacoes_df = associacao_performance.reset_index()
fig_radar = go.Figure()

for i, row in associacoes_df.iterrows():
    fig_radar.add_trace(go.Scatterpolar(
        r=[row['Aproveitamento'], 
           row['Vit√≥ria']/row['Total']*100, 
           row['Empate']/row['Total']*100, 
           row['Derrota']/row['Total']*100],
        theta=['Aproveitamento', 'Vit√≥rias (%)', 'Empates (%)', 'Derrotas (%)'],
        fill='toself',
        name=row['arbitro_associacao']
    ))

fig_radar.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 100]
        )
    ),
    title='Compara√ß√£o de Performance do Farense com Diferentes Associa√ß√µes de √Årbitros',
    showlegend=True
)
fig_radar.write_html("estatisticas_arbitros/radar_associacoes.html")
fig_radar.write_image("estatisticas_arbitros/radar_associacoes.png", width=900, height=600)

# 6. Mapa de calor - Correla√ß√£o entre √°rbitros e resultados
# Criar matriz para o mapa de calor
top_arbitros_names = arbitros_jogos.head(8)['√Årbitro'].tolist()
heatmap_data = df[df['arbitro'].isin(top_arbitros_names)].pivot_table(
    index='arbitro', 
    columns='resultado_farense', 
    aggfunc='size', 
    fill_value=0
)

fig_heatmap = px.imshow(
    heatmap_data, 
    text_auto=True,
    color_continuous_scale='RdYlGn',
    title='Distribui√ß√£o de Resultados por √Årbitro',
    labels=dict(x='Resultado', y='√Årbitro', color='N√∫mero de Jogos')
)
fig_heatmap.write_html("estatisticas_arbitros/heatmap_resultados.html")
fig_heatmap.write_image("estatisticas_arbitros/heatmap_resultados.png", width=900, height=600)

# 7. Gr√°fico temporal de aproveitamento de pontos
aproveitamento_temporal = df.pivot_table(
    index='mes_ano',
    columns='arbitro',
    values='resultado_farense',
    aggfunc=lambda x: (x.value_counts().get('Vit√≥ria', 0) * 3 + x.value_counts().get('Empate', 0)) / (len(x) * 3) * 100
)

# Filtrar apenas √°rbitros com dados suficientes
arbitros_freq = [a for a in aproveitamento_temporal.columns if aproveitamento_temporal[a].count() >= 2]
aproveitamento_temporal = aproveitamento_temporal[arbitros_freq]

fig_temporal = px.line(
    aproveitamento_temporal.reset_index().melt(id_vars='mes_ano'),
    x='mes_ano',
    y='value',
    color='arbitro',
    title='Evolu√ß√£o do Aproveitamento de Pontos do Farense com Diferentes √Årbitros',
    labels={'mes_ano': 'M√™s', 'value': 'Aproveitamento (%)', 'arbitro': '√Årbitro'}
)
fig_temporal.update_layout(yaxis_range=[0, 100])
fig_temporal.write_html("estatisticas_arbitros/aproveitamento_temporal.html")
fig_temporal.write_image("estatisticas_arbitros/aproveitamento_temporal.png", width=900, height=600)

# 8. Distribui√ß√£o de cart√µes amarelos e vermelhos (se houver no JSON)
# Este dado n√£o est√° presente no JSON original, mas mostramos como seria

# 9. Compara√ß√£o entre jogos em casa e fora com os mesmos √°rbitros
casa_fora = df.pivot_table(
    index='arbitro',
    columns='farense_casa',
    values='resultado_farense',
    aggfunc=lambda x: (x.value_counts().get('Vit√≥ria', 0) * 3 + x.value_counts().get('Empate', 0)) / (len(x) * 3) * 100
)

# Renomear colunas para facilitar a compreens√£o
casa_fora.columns = ['Fora', 'Casa']

# Filtrar √°rbitros que apitaram tanto em casa quanto fora
casa_fora = casa_fora.dropna()

if not casa_fora.empty:
    fig_casa_fora = go.Figure()
    fig_casa_fora.add_trace(go.Bar(
        x=casa_fora.index,
        y=casa_fora['Casa'],
        name='Em Casa',
        marker_color='#3498db'
    ))
    fig_casa_fora.add_trace(go.Bar(
        x=casa_fora.index,
        y=casa_fora['Fora'],
        name='Fora',
        marker_color='#e67e22'
    ))
    fig_casa_fora.update_layout(
        title='Aproveitamento do Farense em Casa vs. Fora com os Mesmos √Årbitros',
        xaxis_title='√Årbitro',
        yaxis_title='Aproveitamento (%)',
        barmode='group',
        yaxis_range=[0, 100]
    )
    fig_casa_fora.write_html("estatisticas_arbitros/casa_vs_fora.html")
    fig_casa_fora.write_image("estatisticas_arbitros/casa_vs_fora.png", width=900, height=600)

# 10. Dashboard de estat√≠sticas - Resumo geral
fig_dashboard = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Resultados Gerais da Temporada', 
        'Performance por Associa√ß√£o de √Årbitros',
        'M√©dia de Gols por Jogo', 
        '√Årbitros com Mais Jogos'
    ),
    specs=[
        [{"type": "pie"}, {"type": "bar"}],
        [{"type": "bar"}, {"type": "bar"}]
    ]
)

# 1. Resultados gerais (gr√°fico de pizza)
resultados = df['resultado_farense'].value_counts()
fig_dashboard.add_trace(
    go.Pie(
        labels=resultados.index,
        values=resultados.values,
        marker_colors=['#2ecc71', '#f1c40f', '#e74c3c'],
        textinfo='label+percent',
        hole=0.3
    ),
    row=1, col=1
)

# 2. Performance por associa√ß√£o (top 5)
assoc_top5 = associacao_performance.head(5).reset_index()
fig_dashboard.add_trace(
    go.Bar(
        x=assoc_top5['arbitro_associacao'],
        y=assoc_top5['Aproveitamento'],
        marker_color='#3498db',
        text=assoc_top5['Aproveitamento'].round(1).astype(str) + '%'
    ),
    row=1, col=2
)

# 3. M√©dia de gols por jogo
gols_media = df[['gols_farense', 'gols_sofridos']].mean()
fig_dashboard.add_trace(
    go.Bar(
        x=['Marcados', 'Sofridos'],
        y=gols_media.values,
        marker_color=['#2ecc71', '#e74c3c'],
        text=gols_media.values.round(2)
    ),
    row=2, col=1
)

# 4. √Årbitros com mais jogos (top 5)
top5_arbitros = arbitros_jogos.head(5)
fig_dashboard.add_trace(
    go.Bar(
        x=top5_arbitros['√Årbitro'],
        y=top5_arbitros['Jogos'],
        marker_color='#9b59b6',
        text=top5_arbitros['Jogos']
    ),
    row=2, col=2
)

fig_dashboard.update_layout(
    title_text='Estat√≠sticas do Farense na Temporada 2024/25',
    height=900,
    width=1000,
    showlegend=False
)

fig_dashboard.write_html("estatisticas_arbitros/dashboard_resumo.html")
fig_dashboard.write_image("estatisticas_arbitros/dashboard_resumo.png", width=1000, height=900)

print("\n‚úÖ Gr√°ficos gerados com sucesso no diret√≥rio 'estatisticas_arbitros'")
print("\nResumo de arquivos gerados:")
print("- top10_arbitros_jogos.html/png: Top 10 √°rbitros com mais jogos")
print("- performance_por_arbitro.html/png: Performance do Farense com diferentes √°rbitros")
print("- aproveitamento_por_arbitro.html/png: Aproveitamento de pontos com diferentes √°rbitros")
print("- media_gols_por_arbitro.html/png: M√©dia de gols por jogo com diferentes √°rbitros")
print("- radar_associacoes.html/png: Compara√ß√£o de performance com diferentes associa√ß√µes")
print("- heatmap_resultados.html/png: Distribui√ß√£o de resultados por √°rbitro")
print("- aproveitamento_temporal.html/png: Evolu√ß√£o do aproveitamento ao longo do tempo")
print("- casa_vs_fora.html/png: Compara√ß√£o de aproveitamento em casa vs. fora")
print("- dashboard_resumo.html/png: Dashboard com resumo geral das estat√≠sticas")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.zerozero.pt/jogo/2025-05-11-vitoria-sc-farense/10240205"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Selecionar div info
info_div = soup.select_one("#page_header > div > div.info")

if not info_div:
    print("Div 'info' n√£o encontrada.")
    exit()

# Extrair dados principais
data = info_div.select_one(".header i.fa-calendar").next_sibling.strip()
hora = info_div.select_one(".header i.fa-clock").next_sibling.strip()
estadio_link = info_div.select_one(".header a[href^='/estadio']")
estadio = estadio_link.text.strip() if estadio_link else "N/A"
local = estadio_link.next_sibling.strip(" ()") if estadio_link and estadio_link.next_sibling else "N/A"
publico = info_div.select_one(".header i.fa-people-group")
publico = publico.next_sibling.strip() if publico else "N/A"
tv = info_div.select_one(".header i.fa-tv")
tv = tv.next_sibling.strip() if tv else "N/A"
arbitro_link = info_div.select_one(".header a[href^='/arbitro']")
arbitro = arbitro_link.text.strip() if arbitro_link else "N/A"
campeonato_link = info_div.select_one(".header a[href^='/edicao']")
campeonato = campeonato_link.text.strip() if campeonato_link else "N/A"
jornada_text = info_div.select_one(".header").text
import re
jornada_match = re.search(r'Jornada\s+(\d+)', jornada_text)
jornada = jornada_match.group(1) if jornada_match else "N/A"

# Equipas e resultado
home_team = info_div.select_one(".home .team a").text.strip()
away_team = info_div.select_one(".away .team a").text.strip()
resultado = info_div.select_one(".score a").text.strip()

# Intervalo (se existir)
intervalo_span = info_div.select_one("div.content > div.score > span")
if intervalo_span:
    intervalo_text = intervalo_span.get_text(strip=True)
    # intervalo_text ser√° algo como "45': Intervalo: 0-1"
    # Podemos extrair s√≥ a parte do resultado do intervalo
    import re
    match = re.search(r'Intervalo:\s*([\d\-]+)', intervalo_text)
    intervalo = match.group(1) if match else "N/A"
else:
    intervalo = "N/A"

# Marcadores - equipa casa
home_scorers = info_div.select(".home .scorers span span span")
golos_casa = []
for s in home_scorers:
    tempo_element = s.select_one(".time")
    jogador_element = s.select_one("a")
    if tempo_element and jogador_element:
        tempo = tempo_element.text.strip()
        jogador = jogador_element.text.strip()
        golos_casa.append(f"{tempo}' {jogador}")

# Marcadores - equipa fora
away_scorers = info_div.select(".away .scorers span span span")
golos_fora = []
for s in away_scorers:
    tempo_element = s.select_one(".time")
    jogador_element = s.select_one("a")
    if tempo_element and jogador_element:
        tempo = tempo_element.text.strip()
        jogador = jogador_element.text.strip()
        golos_fora.append(f"{tempo}' {jogador}")

# Apresenta√ß√£o limpa no ecr√£
print(f"Data: {data}")
print(f"Hora: {hora}")
print(f"Est√°dio: {estadio} ({local})")
print(f"P√∫blico: {publico}")
print(f"Transmiss√£o TV: {tv}")
print(f"√Årbitro: {arbitro}")
print(f"Campeonato: {campeonato}")
print(f"Jornada: {jornada}")
print()
print(f"Equipa Casa: {home_team}")
print(f"Equipa Fora: {away_team}")
print(f"Resultado Final: {resultado}")
print(f"Intervalo: {intervalo}")
print()
print(f"Golos {home_team}:")
if golos_casa:
    for g in golos_casa:
        print(f"  - {g}")
else:
    print("  Nenhum")

print(f"\nGolos {away_team}:")
if golos_fora:
    for g in golos_fora:
        print(f"  - {g}")
else:
    print("  Nenhum")
```
