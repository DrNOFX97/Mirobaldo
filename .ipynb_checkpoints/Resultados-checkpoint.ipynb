# Calendário de Jogos SC Farense - Exportação de Dados

Este projeto tem como objetivo obter e exportar o calendário de jogos do SC Farense a partir do site ZeroZero.pt, organizando e exportando os dados para diferentes formatos, como Excel, CSV, TXT e Word.

## Funcionalidades

1. **Calendário completo de jogos:** Lista todos os jogos do SC Farense para as épocas disponíveis, mostrando as informações como data, hora, local, equipa adversária, resultado e jornada.
2. **Calendário por competição:** Filtra os jogos por competição específica.
3. **Exportação de dados:**
   - Exporta os dados para **Excel** (por competição ou completo).
   - Exporta todos os jogos para um único arquivo **CSV**.
   - Exporta todos os jogos para um arquivo de texto **TXT** formatado.
   - Exporta todos os jogos para um documento **Word**.

## Requisitos

- Python 3.x
- Bibliotecas necessárias:
  - `pandas`
  - `beautifulsoup4`
  - `requests`
  - `tqdm`
  - `tabulate`
  - `python-docx`

```bash
pip install pandas beautifulsoup4 requests tqdm tabulate python-docx
```

## Detalhes das Funcionalidades

### Menu

O menu apresenta as seguintes opções para interagir com os dados:

1. **Calendário completo de jogos:** Apresenta uma tabela organizada de todos os jogos por ordem de data.
2. **Calendário por competição:** Permite selecionar uma competição específica para visualizar os jogos.
3. **Exportar dados para Excel:** Permite exportar todos os jogos ou apenas de uma competição específica para arquivos Excel.
4. **Exportar todos os dados para CSV:** Exporta todos os jogos para um único arquivo CSV.
5. **Exportar todos os dados para TXT:** Gera um arquivo de texto com os jogos formatados de forma legível.
6. **Exportar todos os dados para Word:** Exporta os dados para um documento Word com tabelas organizadas.

### Exportar para Excel

Você pode exportar os dados para uma planilha Excel de uma competição específica ou de todas as competições. O arquivo será salvo na pasta local com o nome da competição.

### Exportar para CSV

Exporta todos os jogos de todas as competições para um único arquivo CSV chamado `todos_os_jogos.csv`.

### Exportar para TXT

Exporta todos os jogos para um arquivo de texto formatado, onde cada jogo é organizado de forma tabular.

### Exportar para Word

Gera um documento Word chamado `todos_os_jogos.docx`, onde os jogos são organizados em tabelas, uma para cada competição.

## Estrutura do Código

- **menu:** Função principal que exibe as opções do menu e captura a escolha do utilizador.
- **listar_epocas:** Mostra as épocas disponíveis para seleção.
- **calendario_completo:** Carrega e exibe o calendário completo dos jogos.
- **exportar_para_excel:** Menu para exportar dados para arquivos Excel.
- **exportar_tudo_para_excel:** Exporta todos os jogos de todas as competições para arquivos Excel.
- **exportar_competicao_especifica_para_excel:** Exporta os jogos de uma competição específica para um arquivo Excel.
- **exportar_tudo_para_csv:** Exporta todos os jogos para um arquivo CSV.
- **exportar_tudo_para_txt:** Exporta todos os jogos para um arquivo TXT.
- **exportar_tudo_para_word:** Exporta todos os jogos para um documento Word.
- **mostrar_dados_em_txt:** Exibe os dados dos jogos no formato tabular diretamente no Jupyter Notebook.

````python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm.auto import tqdm
from tabulate import tabulate
from IPython.display import display, Markdown
from docx import Document
from docx.shared import Cm

# Criando dicionário de épocas
epocas = {}
for i in range(63, 156):
    epoca = i - 62
    ano_inicio = 1933 + epoca
    ano_fim = ano_inicio + 1
    epocas[i] = f"{ano_inicio}/{str(ano_fim)[-2:]}"

base_url = 'https://www.zerozero.pt/equipa/farense/10/jogos?grp=1&epoca_id='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

jogos = []
competicoes = {}

# Contando épocas disponíveis
total_epocas = len(epocas)

display(Markdown(f"\nCarregando os dados..."))
display(Markdown(f"Aguarde cerca de um minuto... \nEstamos a recolher os dados de {total_epocas} épocas, seja paciente"))

# Usando tqdm para mostrar a linha de progresso
for epoca_id, epoca_nome in tqdm(epocas.items(), desc="Progresso", unit=" época"):

    url = f'{base_url}{epoca_id}'

    try:
        response = requests.get(url, headers=headers)
        # Detecta automaticamente a codificação e a define
        response.encoding = response.apparent_encoding

        # Parsing HTML
        soup = BeautifulSoup(response.content, 'html.parser')
        linhas = soup.find_all('tr', class_='parent')

        mapeamento_competicao = {
            'AF Algarve': 'Associação',
            'Taça Futsal': 'Competição',
            'Jun.C S15': 'Escalão',
            '1ªF': 'Fase',
            'GB': 'Grupo',
            '23/24': 'Época'
        }

        for linha in linhas:
            celulas = linha.find_all('td')
            if len(celulas) >= 8:
                competicao = celulas[7].text.strip()
                resultado = celulas[0].text.strip()
                if resultado == "h2h":
                    resultado = "-"
                local = celulas[3].text.strip()
                if local == '(F)':
                    local = 'Fora'
                elif local == '(C)':
                    local = 'Casa'
                jogo = {
                    'V-E-D': resultado,
                    'Data': celulas[1].text.strip(),
                    'Hora': celulas[2].text.strip(),
                    'Local': local,
                    'Equipa': celulas[5].text.strip(),
                    'Resultado': celulas[6].text.strip(),
                    'Jornada': celulas[8].text.strip()
                }
                jogos.append(jogo)

                partes_competicao = competicao.split()
                competicao_decodificada = {}
                for parte in partes_competicao:
                    if parte in mapeamento_competicao:
                        competicao_decodificada[mapeamento_competicao[parte]] = parte
                    else:
                        competicao_decodificada['Outro'] = parte

                if competicao in competicoes:
                    competicoes[competicao]['Jogos'].append(jogo)
                else:
                    competicoes[competicao] = {
                        'Texto Original': competicao,
                        'Decodificado': competicao_decodificada,
                        'Jogos': [jogo]
                    }
    except (requests.RequestException, ValueError) as e:
        display(Markdown(f"Erro ao processar época {epoca_id}: {e}"))

def calendario_por_competicao(competicoes):
    display(Markdown("### Selecione uma competição:"))
    for idx, competicao in enumerate(competicoes.keys(), 1):
        display(Markdown(f"{idx}. {competicao}"))
    
    escolha = int(input("Escolha o número da competição: ")) - 1
    competicao_escolhida = list(competicoes.keys())[escolha]

    display(Markdown(f"### Calendário para a competição {competicao_escolhida}:"))
    df = pd.DataFrame(competicoes[competicao_escolhida]['Jogos'])
    df = df.sort_values(by='Data', ascending=True)
    display(Markdown(tabulate(df, headers='keys', tablefmt='pipe', showindex=False)))

def calendario_completo(competicoes):
    display(Markdown(f"### Calendário completo de todas as competições"))
    for competicao, dados in competicoes.items():
        display(Markdown(f"#### {competicao}"))
        df = pd.DataFrame(dados['Jogos'])
        df = df.sort_values(by='Data', ascending=True)
        display(Markdown(tabulate(df, headers='keys', tablefmt='pipe', showindex=False)))

def exportar_para_excel(competicoes):
    writer = pd.ExcelWriter('calendario_farense.xlsx', engine='xlsxwriter')
    for competicao, dados in competicoes.items():
        df = pd.DataFrame(dados['Jogos'])
        df.to_excel(writer, sheet_name=competicao[:31], index=False)
    writer.save()
    display(Markdown("Dados exportados para Excel com sucesso!"))

def exportar_tudo_para_csv(competicoes):
    for competicao, dados in competicoes.items():
        df = pd.DataFrame(dados['Jogos'])
        df.to_csv(f'calendario_{competicao}.csv', index=False)
    display(Markdown("Dados exportados para CSV com sucesso!"))

def exportar_tudo_para_txt(competicoes):
    with open('calendario_farense.txt', 'w') as f:
        for competicao, dados in competicoes.items():
            f.write(f"Competição: {competicao}\n")
            f.write(tabulate(pd.DataFrame(dados['Jogos']), headers='keys', tablefmt='pipe', showindex=False))
            f.write("\n\n")
    display(Markdown("Dados exportados para TXT com sucesso!"))

def exportar_tudo_para_word(competicoes):
    doc = Document()
    doc.add_heading('Calendário de Jogos - Farense', 0)
    for competicao, dados in competicoes.items():
        doc.add_heading(competicao, level=1)
        tabela = doc.add_table(rows=1, cols=7)
        hdr_cells = tabela.rows[0].cells
        hdr_cells[0].text = 'V-E-D'
        hdr_cells[1].text = 'Data'
        hdr_cells[2].text = 'Hora'
        hdr_cells[3].text = 'Local'
        hdr_cells[4].text = 'Equipa'
        hdr_cells[5].text = 'Resultado'
        hdr_cells[6].text = 'Jornada'
        for jogo in dados['Jogos']:
            row_cells = tabela.add_row().cells
            row_cells[0].text = jogo['V-E-D']
            row_cells[1].text = jogo['Data']
            row_cells[2].text = jogo['Hora']
            row_cells[3].text = jogo['Local']
            row_cells[4].text = jogo['Equipa']
            row_cells[5].text = jogo['Resultado']
            row_cells[6].text = jogo['Jornada']
    doc.save('calendario_farense.docx')
    display(Markdown("Dados exportados para Word com sucesso!"))

# Função para exibir o menu
def menu():
    display(Markdown("### Escolha uma opção:"))
    display(Markdown("1. Calendário completo de jogos"))
    display(Markdown("2. Calendário por competição"))
    display(Markdown("3. Exportar dados para Excel"))
    display(Markdown("4. Exportar todos os dados para um único CSV"))
    display(Markdown("5. Exportar todos os dados para um único TXT"))
    display(Markdown("6. Exportar todos os dados para um documento Word"))
    display(Markdown("7. Sair"))
    escolha = input("Digite o número da opção desejada: ")
    return escolha

# Menu principal
escolha = None
while escolha != '7':
    escolha = menu()
    if escolha == '1':
        calendario_completo(competicoes)
    elif escolha == '2':
        calendario_por_competicao(competicoes)
    elif escolha == '3':
        exportar_para_excel(competicoes)
    elif escolha == '4':
        exportar_tudo_para_csv(competicoes)
    elif escolha == '5':
        exportar_tudo_para_txt(competicoes)
    elif escolha == '6':
        exportar_tudo_para_word(competicoes)
    elif escolha == '7':
        display(Markdown("Saindo..."))
    else:
        display(Markdown("Opção inválida. Por favor, escolha uma opção válida."))
````

````python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm.auto import tqdm
from tabulate import tabulate
from IPython.display import display, Markdown
import re

# Criando dicionário de épocas
epocas = {}
for i in range(63, 156):
    epoca = i - 63
    ano_inicio = 1933 + epoca
    ano_fim = ano_inicio + 1
    epocas[i] = f"{ano_inicio}-{str(ano_fim)[-2:]}"

base_url = 'https://www.zerozero.pt/equipa/farense/10/jogos?grp=1&epoca_id='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Função para corrigir caracteres especiais e limpar texto
def corrigir_caracteres_especiais(texto):
    # Substitui caracteres problemáticos por equivalentes
    texto_corrigido = texto.replace('ZS', 'Zona Sul').replace('ZD', 'Zona D').replace('Campees', 'Campeões').replace('Taa', 'Taça')
    # Remove espaços em branco extras e caracteres indesejados
    texto_corrigido = re.sub(r'\s+', ' ', texto_corrigido).strip()
    return texto_corrigido

# Função para exibir uma tabela formatada com os jogos
def exibir_tabela_jogos(competicao, epoca_nome, jogos):
    competicao = corrigir_caracteres_especiais(competicao)
    display(Markdown(f"---\n### Época {epoca_nome}"))
    display(Markdown(f"---\n### {competicao}"))
    df = pd.DataFrame(jogos)
    df = df.sort_values(by='Data', ascending=True)
    display(Markdown(tabulate(df, headers='keys', tablefmt='pipe', showindex=False)))

# Usando tqdm para mostrar a linha de progresso
for epoca_id, epoca_nome in tqdm(epocas.items(), desc="Processando épocas", unit="época"):
    url = f'{base_url}{epoca_id}'

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        # Decodificando conteúdo com tratamento de erros
        response.encoding = response.apparent_encoding  # Garantir a codificação correta
        soup = BeautifulSoup(response.text, 'html.parser')

        # Encontrar todas as linhas relevantes na tabela
        linhas = soup.find_all('tr', class_='parent')

        competicoes = {}  # Reiniciar competições para cada época

        for linha in linhas:
            celulas = linha.find_all('td')
            if len(celulas) >= 8:
                competicao = celulas[7].text.strip()
                resultado = celulas[0].text.strip()
                if resultado == "h2h":
                    resultado = "-"
                local = celulas[3].text.strip()
                if local == '(F)':
                    local = 'Fora'
                elif local == '(C)':
                    local = 'Casa'
                
                jogo = {
                    'V-E-D': resultado,
                    'Data': celulas[1].text.strip(),
                    'Hora': celulas[2].text.strip(),
                    'Local': local,
                    'Equipa': celulas[5].text.strip(),
                    'Resultado': celulas[6].text.strip(),
                    'Jornada': celulas[8].text.strip()
                }

                if competicao in competicoes:
                    competicoes[competicao].append(jogo)
                else:
                    competicoes[competicao] = [jogo]

        # Exibir os jogos para cada competição, assim que os dados forem obtidos
        if competicoes:
            for competicao, jogos in competicoes.items():
                exibir_tabela_jogos(competicao, epoca_nome, jogos)
        else:
            display(Markdown(f"**Sem jogos encontrados para a época {epoca_nome}.**"))

    except requests.RequestException as e:
        display(Markdown(f"Erro ao acessar {url}: {e}"))

display(Markdown("### Processamento completo!"))
````

````python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm.auto import tqdm
from tabulate import tabulate
from IPython.display import display, Markdown
from docx import Document
from docx.shared import Cm

# Criando dicionário de épocas
epocas = {}
for i in range(63, 156):
    epoca = i - 62
    ano_inicio = 1933 + epoca
    ano_fim = ano_inicio + 1
    epocas[i] = f"{ano_inicio}-{str(ano_fim)[-2:]}"

base_url = 'https://www.zerozero.pt/equipa/farense/10/jogos?grp=1&epoca_id='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

jogos = []
competicoes = {}

# Contando épocas disponíveis
total_epocas = len(epocas)

display(Markdown(f"\nCarregando os dados..."))
display(Markdown(f"Aguarde cerca de um minuto... \nEstamos a recolher os dados de {total_epocas} épocas, seja paciente"))

# Usando tqdm para mostrar a linha de progresso
for epoca_id, epoca_nome in tqdm(epocas.items(), desc="Progresso", unit=" época"):

    url = f'{base_url}{epoca_id}'

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        soup = BeautifulSoup(response.content, 'html.parser')

        mapeamento_competicao = {
            'AF Algarve': 'Associação',
            'Taça Futsal': 'Competição',
            'Jun.C S15': 'Escalão',
            '1ªF': 'Fase',
            'GB': 'Grupo',
            '23/24': 'Época'
        }

        linhas = soup.find_all('tr', class_='parent')

        for linha in linhas:
            celulas = linha.find_all('td')
            if len(celulas) >= 8:
                competicao = celulas[7].text.strip()
                resultado = celulas[0].text.strip()
                if resultado == "h2h":
                    resultado = "-"
                local = celulas[3].text.strip()
                if local == '(F)':
                    local = 'Fora'
                elif local == '(C)':
                    local = 'Casa'
                jogo = {
                    'V-E-D': resultado,
                    'Data': celulas[1].text.strip(),
                    'Hora': celulas[2].text.strip(),
                    'Local': local,
                    'Equipa': celulas[5].text.strip(),
                    'Resultado': celulas[6].text.strip(),
                    'Jornada': celulas[8].text.strip()
                }
                jogos.append(jogo)

                partes_competicao = competicao.split()
                competicao_decodificada = {}
                for parte in partes_competicao:
                    if parte in mapeamento_competicao:
                        competicao_decodificada[mapeamento_competicao[parte]] = parte
                    else:
                        competicao_decodificada['Outro'] = parte

                if competicao in competicoes:
                    competicoes[competicao]['Jogos'].append(jogo)
                else:
                    competicoes[competicao] = {
                        'Texto Original': competicao,
                        'Decodificado': competicao_decodificada,
                        'Jogos': [jogo]
                    }
    except (requests.RequestException, ValueError) as e:
        display(Markdown(f"Erro ao processar época {epoca_id}: {e}"))

# Função para exibir menu de opções
def menu():
    display(Markdown("### Escolha uma opção:"))
    display(Markdown("1. Calendário completo de jogos"))
    display(Markdown("2. Calendário por competição"))
    display(Markdown("3. Exportar dados para Excel"))
    display(Markdown("4. Exportar todos os dados para um único CSV"))
    display(Markdown("5. Exportar todos os dados para um único TXT"))
    display(Markdown("6. Exportar todos os dados para um documento Word"))
    display(Markdown("7. Sair"))
    escolha = input("Digite o número da opção desejada: ")
    return escolha

# Função para listar as épocas disponíveis
def listar_epocas():
    display(Markdown("### Lista de épocas disponíveis:"))
    for epoca_id, epoca_nome in epocas.items():
        display(Markdown(f"{epoca_id}. Época {epoca_nome}"))

# Função para mostrar o calendário completo de jogos
def calendario_completo(competicoes):
    display(Markdown("### Carregando calendário completo..."))
    for competicao, dados in tqdm(competicoes.items(), desc="Progresso", unit="competição"):
        display(Markdown("---"))
        display(Markdown(f"### Época {epoca_nome}"))
        display(Markdown("---"))
        display(Markdown(f"### {competicao}"))
        df = pd.DataFrame(dados['Jogos'])
        df = df.sort_values(by='Data', ascending=True)
        display(Markdown(tabulate(df, headers='keys', tablefmt='pipe', showindex=False)))

# Função para exportar dados para Excel
def exportar_para_excel(competicoes):
    display(Markdown("### Menu de Exportação para Excel:"))
    display(Markdown("1. Exportar Tudo"))
    display(Markdown("2. Exportar Competição Específica"))
    opcao = input("Escolha a opção desejada (1/2): ")

    if opcao == '1':
        exportar_tudo_para_excel(competicoes)
    elif opcao == '2':
        exportar_competicao_especifica_para_excel(competicoes)
    else:
        display(Markdown("Opção inválida."))

def exportar_tudo_para_excel(competicoes):
    opcao = input("\nDeseja exportar os dados de todas as competições para Excel? (S/N): ").upper()
    if opcao == 'S':
        display(Markdown("### Exportando dados de todas as competições para Excel..."))
        for competicao, dados in tqdm(competicoes.items(), desc="Progresso", unit="competição"):
            display(Markdown(f"Exportando dados da competição: {competicao}"))
            df = pd.DataFrame(dados['Jogos'])
            nome_planilha = competicao[:31].replace('[', '-').replace(']', '-').replace(':', '-').replace('*', '-').replace('?', '-').replace('/', '-').replace('\\', '-')
            df.to_excel(f'{nome_planilha}.xlsx', index=False)
            display(Markdown(f"Dados exportados com sucesso para a planilha '{nome_planilha}.xlsx'"))
    else:
        display(Markdown("Exportação para Excel cancelada."))

def exportar_competicao_especifica_para_excel(competicoes):
    display(Markdown("### Lista de Competições Disponíveis:"))
    for idx, competicao in enumerate(competicoes.keys(), 1):
        display(Markdown(f"{idx}. {competicao}"))

    display(Markdown(f"{len(competicoes) + 1}. Voltar atrás"))
    escolha = int(input("Selecione o número da competição: "))

    if escolha == len(competicoes) + 1:
        return

    competicao_escolhida = list(competicoes.keys())[escolha - 1]

    display(Markdown(f"### Exportando dados da competição '{competicao_escolhida}' para Excel..."))
    df = pd.DataFrame(competicoes[competicao_escolhida]['Jogos'])
    nome_planilha = competicao_escolhida[:31].replace('[', '-').replace(']', '-').replace(':', '-').replace('*', '-').replace('?', '-').replace('/', '-').replace('\\', '-')
    df.to_excel(f'{nome_planilha}.xlsx', index=False)
    display(Markdown(f"Dados exportados com sucesso para a planilha '{nome_planilha}.xlsx'"))

# Função para exportar dados para CSV
def exportar_tudo_para_csv(competicoes):
    opcao = input("\nDeseja exportar os dados de todas as competições para um único CSV? (S/N): ").upper()
    if opcao == 'S':
        display(Markdown("### Exportando dados de todas as competições para um único arquivo CSV..."))
        todos_jogos = []
        for competicao, dados in tqdm(competicoes.items(), desc="Progresso", unit="competição"):
            for jogo in dados['Jogos']:
                jogo['Competicao'] = competicao
                todos_jogos.append(jogo)
        df = pd.DataFrame(todos_jogos)
        df.to_csv('todos_os_jogos.csv', index=False)
        display(Markdown("Dados exportados com sucesso para o arquivo 'todos_os_jogos.csv'"))
    else:
        display(Markdown("Exportação para CSV cancelada."))

# Função para exportar dados para TXT
def exportar_tudo_para_txt(competicoes):
    opcao = input("\nDeseja exportar os dados de todas as competições para um único TXT? (S/N): ").upper()
    if opcao == 'S':
        display(Markdown("### Exportando dados de todas as competições para um único arquivo TXT..."))
        with open('todos_os_jogos.txt', 'w') as file:
            for competicao, dados in tqdm(competicoes.items(), desc="Progresso", unit="competição"):
                file.write(f"---\nCompetição: {competicao}\n")
                for jogo in dados['Jogos']:
                    file.write(f"{jogo}\n")
        display(Markdown("Dados exportados com sucesso para o arquivo 'todos_os_jogos.txt'"))
    else:
        display(Markdown("Exportação para TXT cancelada."))

# Função para exportar dados para Word
def exportar_para_word(competicoes):
    opcao = input("\nDeseja exportar os dados de todas as competições para um documento Word? (S/N): ").upper()
    if opcao == 'S':
        display(Markdown("### Exportando dados para Word..."))
        doc = Document()
        doc.add_heading('Jogos SC Farense - Todas as Competições', 0)
        for competicao, dados in tqdm(competicoes.items(), desc="Progresso", unit="competição"):
            doc.add_heading(competicao, level=1)
            for jogo in dados['Jogos']:
                doc.add_paragraph(f"{jogo['Data']} - {jogo['Hora']} - {jogo['Equipa']} - {jogo['Resultado']} - {jogo['Local']}")
        doc.save('todos_os_jogos.docx')
        display(Markdown("Dados exportados com sucesso para o documento 'todos_os_jogos.docx'"))
    else:
        display(Markdown("Exportação para Word cancelada."))

# Menu principal
while True:
    opcao = menu()
    if opcao == '1':
        calendario_completo(competicoes)
    elif opcao == '2':
        display(Markdown("### Escolha a competição:"))
        for idx, competicao in enumerate(competicoes.keys(), 1):
            display(Markdown(f"{idx}. {competicao}"))
        escolha = int(input("Digite o número da competição desejada: "))
        competicao_escolhida = list(competicoes.keys())[escolha - 1]
        calendario_completo({competicao_escolhida: competicoes[competicao_escolhida]})
    elif opcao == '3':
        exportar_para_excel(competicoes)
    elif opcao == '4':
        exportar_tudo_para_csv(competicoes)
    elif opcao == '5':
        exportar_tudo_para_txt(competicoes)
    elif opcao == '6':
        exportar_para_word(competicoes)
    elif opcao == '7':
        display(Markdown("A sair..."))
        break
    else:
        display(Markdown("Opção inválida. Por favor, tente novamente."))
````

````python
import aiohttp
import asyncio
from bs4 import BeautifulSoup
from tqdm.auto import tqdm
from IPython.display import display, Markdown

# Criando dicionário de épocas
epocas = {}
for i in range(63, 156):
    epoca = i - 62
    ano_inicio = 1933 + epoca
    ano_fim = ano_inicio + 1
    epocas[i] = f"{ano_inicio}/{str(ano_fim)[-2:]}"

base_url = 'https://www.zerozero.pt/equipa/farense/10/jogos?grp=1&epoca_id='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

jogos = []
competicoes = {}

# Mensagens de progresso
display(Markdown(f"\nCarregando os dados..."))
display(Markdown(f"Aguarde cerca de um minuto... \nEstamos a recolher os dados de {len(epocas)} épocas, seja paciente"))

# Função para coletar dados de uma única época
async def coleta_dados_epoca(session, epoca_id, epoca_nome):
    url = f'{base_url}{epoca_id}'
    try:
        async with session.get(url) as response:
            response_text = await response.text()
            soup = BeautifulSoup(response_text, 'html.parser')
            linhas = soup.find_all('tr', class_='parent')

            mapeamento_competicao = {
                'AF Algarve': 'Associação',
                'Taça Futsal': 'Competição',
                'Jun.C S15': 'Escalão',
                '1ªF': 'Fase',
                'GB': 'Grupo',
                '23/24': 'Época'
            }

            for linha in linhas:
                celulas = linha.find_all('td')
                if len(celulas) >= 8:
                    competicao = celulas[7].text.strip()
                    resultado = celulas[0].text.strip()
                    if resultado == "h2h":
                        resultado = "-"
                    local = celulas[3].text.strip()
                    if local == '(F)':
                        local = 'Fora'
                    elif local == '(C)':
                        local = 'Casa'

                    # Obter a imagem da competição ou do jogo
                    img_tag = celulas[4].find('img')
                    img_url = img_tag['src'] if img_tag and 'src' in img_tag.attrs else None
                    if img_url and img_url.startswith('/'):
                        img_url = f'https://www.zerozero.pt{img_url}'  # Completar a URL base

                    # Criar o dicionário do jogo
                    jogo = {
                        'V-E-D': resultado,
                        'Data': celulas[1].text.strip(),
                        'Hora': celulas[2].text.strip(),
                        'Local': local,
                        'Equipa': celulas[5].text.strip(),
                        'Resultado': celulas[6].text.strip(),
                        'Jornada': celulas[8].text.strip(),
                        'Imagem': img_url
                    }
                    jogos.append(jogo)

                    partes_competicao = competicao.split()
                    competicao_decodificada = {}
                    for parte in partes_competicao:
                        if parte in mapeamento_competicao:
                            competicao_decodificada[mapeamento_competicao[parte]] = parte
                        else:
                            competicao_decodificada['Outro'] = parte

                    if competicao in competicoes:
                        competicoes[competicao]['Jogos'].append(jogo)
                    else:
                        competicoes[competicao] = {
                            'Texto Original': competicao,
                            'Decodificado': competicao_decodificada,
                            'Jogos': [jogo]
                        }
    except Exception as e:
        display(Markdown(f"Erro ao processar época {epoca_nome}: {e}"))

# Função principal para fazer a coleta de dados assíncrona
async def coleta_dados():
    async with aiohttp.ClientSession(headers=headers) as session:
        tasks = []
        for epoca_id, epoca_nome in epocas.items():
            task = coleta_dados_epoca(session, epoca_id, epoca_nome)
            tasks.append(task)
        
        # Executar todas as tarefas de forma paralela
        for f in tqdm(asyncio.as_completed(tasks), total=len(tasks), desc="Progresso", unit=" época"):
            await f

# Iniciar a coleta de dados assíncrona
await coleta_dados()

# Exibir os jogos com imagens
for jogo in jogos:
    img_html = f'<img src="{jogo["Imagem"]}" width="30" height="30" style="vertical-align: middle;">' if jogo["Imagem"] else ""
    display(Markdown(f"{jogo['Data']} - {img_html} {jogo['Equipa']} | {jogo['V-E-D']} | {jogo['Local']} | {jogo['Resultado']} | {jogo['Jornada']}"))
````

````python
import pandas as pd
from bs4 import BeautifulSoup
import aiohttp
import asyncio
import nest_asyncio
from tqdm.auto import tqdm
from IPython.display import display, Markdown
import requests

# Permitir loops aninhados no Jupyter
nest_asyncio.apply()

# Criando dicionário de épocas
epocas = {}
for i in range(63, 156):
    epoca = i - 62
    ano_inicio = 1933 + epoca
    ano_fim = ano_inicio + 1
    epocas[i] = f"{ano_inicio}/{str(ano_fim)[-2:]}"

base_url = 'https://www.zerozero.pt/equipa/farense/10/jogos?grp=1&epoca_id='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

jogos = []
competicoes = {}

# Função assíncrona para obter dados de uma época
async def obter_dados_epoca(session, epoca_id, epoca_nome):
    url = f'{base_url}{epoca_id}'
    try:
        async with session.get(url) as response:
            content = await response.text()
            soup = BeautifulSoup(content, 'html.parser')
            linhas = soup.find_all('tr', class_='parent')

            mapeamento_competicao = {
                'AF Algarve': 'Associação',
                'Taça Futsal': 'Competição',
                'Jun.C S15': 'Escalão',
                '1ªF': 'Fase',
                'GB': 'Grupo',
                '23/24': 'Época'
            }

            for linha in linhas:
                celulas = linha.find_all('td')
                if len(celulas) >= 8:
                    competicao = celulas[7].text.strip()
                    resultado = celulas[0].text.strip()
                    if resultado == "h2h":
                        resultado = "-"
                    local = celulas[3].text.strip()
                    if local == '(F)':
                        local = 'Fora'
                    elif local == '(C)':
                        local = 'Casa'

                    # Obter a imagem da competição ou do jogo
                    img_tag = celulas[4].find('img')
                    img_url = img_tag['src'] if img_tag and 'src' in img_tag.attrs else None
                    if img_url and img_url.startswith('/'):
                        img_url = f'https://www.zerozero.pt{img_url}'

                    # Criar o dicionário do jogo
                    jogo = {
                        'V-E-D': resultado,
                        'Data': celulas[1].text.strip(),
                        'Hora': celulas[2].text.strip(),
                        'Local': local,
                        'Imagem': img_url,  # Adicionando a imagem ao dicionário do jogo
                        'Equipa': celulas[5].text.strip(),
                        'Resultado': celulas[6].text.strip(),
                        'Jornada': celulas[8].text.strip(),
                        'Época': epoca_nome  # Adicionando a época
                    }
                    jogos.append(jogo)

                    partes_competicao = competicao.split()
                    competicao_decodificada = {}
                    for parte in partes_competicao:
                        if parte in mapeamento_competicao:
                            competicao_decodificada[mapeamento_competicao[parte]] = parte
                        else:
                            competicao_decodificada['Outro'] = parte

                    if competicao in competicoes:
                        competicoes[competicao]['Jogos'].append(jogo)
                    else:
                        competicoes[competicao] = {
                            'Texto Original': competicao,
                            'Decodificado': competicao_decodificada,
                            'Jogos': [jogo]
                        }
    except (aiohttp.ClientError, ValueError) as e:
        display(Markdown(f"Erro ao processar época {epoca_id}: {e}"))

# Função para executar as tarefas assíncronas em paralelo
async def coletar_dados():
    async with aiohttp.ClientSession(headers=headers) as session:
        tasks = []
        for epoca_id, epoca_nome in epocas.items():
            task = obter_dados_epoca(session, epoca_id, epoca_nome)
            tasks.append(task)

        # Executa as tarefas de forma paralela e acompanha o progresso
        for task in tqdm(asyncio.as_completed(tasks), total=len(tasks), desc="Progresso", unit=" época"):
            await task

# Função para iniciar a coleta de dados
def iniciar_coleta_dados():
    display(Markdown(f"\nCarregando os dados..."))
    display(Markdown(f"Aguarde cerca de um minuto... \nEstamos a recolher os dados de {len(epocas)} épocas, seja paciente"))
    asyncio.run(coletar_dados())

# Função para exibir o calendário completo organizado por épocas com imagens
def calendario_completo_organizado_por_epocas(jogos):
    display(Markdown(f"### Calendário completo organizado por épocas"))

    # Organizar jogos por épocas
    jogos_por_epoca = {}
    for jogo in jogos:
        epoca = jogo['Época']
        if epoca not in jogos_por_epoca:
            jogos_por_epoca[epoca] = []
        jogos_por_epoca[epoca].append(jogo)
    
    # Exibir as tabelas organizadas por época
    for epoca, jogos_da_epoca in sorted(jogos_por_epoca.items()):
        display(Markdown(f"#### Época {epoca}"))
        
        # Criar uma lista de jogos com as imagens embutidas
        jogos_com_imagens = []
        for jogo in jogos_da_epoca:
            # Adicionando a tag <img> diretamente à tabela, antes da equipa
            img_html = f'<img src="{jogo["Imagem"]}" width="30">' if jogo['Imagem'] else ''
            jogos_com_imagens.append([
                jogo['V-E-D'], 
                jogo['Data'], 
                jogo['Hora'], 
                jogo['Local'], 
                img_html,  # Adicionando imagem antes da equipa
                jogo['Equipa'], 
                jogo['Resultado'], 
                jogo['Jornada']
            ])
        
        # Convertendo a lista em um DataFrame
        df = pd.DataFrame(jogos_com_imagens, columns=['V-E-D', 'Data', 'Hora', 'Local', 'Imagem', 'Equipa', 'Resultado', 'Jornada'])
        df = df.sort_values(by='Data', ascending=True)

        # Exibindo a tabela com imagens
        display(Markdown(df.to_html(escape=False, index=False)))

# Função principal
def menu():
    display(Markdown("### Escolha uma opção:"))
    display(Markdown("1. Iniciar coleta de dados"))
    display(Markdown("2. Exibir calendário completo organizado por épocas"))
    display(Markdown("3. Sair"))

    escolha = input("Digite o número da opção desejada: ")
    return escolha

# Loop principal
escolha = None
while escolha != '3':
    escolha = menu()
    if escolha == '1':
        iniciar_coleta_dados()
    elif escolha == '2':
        calendario_completo_organizado_por_epocas(jogos)
    elif escolha == '3':
        display(Markdown("Saindo..."))
    else:
        display(Markdown("Opção inválida. Por favor, escolha uma opção válida."))
````
